```python
#rawdata berisi tabel dengan kolom yang diuraikan
rawdata = df[['dt_iso', 'city_name', 'temp', 'pressure', 'humidity', 'wind_speed', 'wind_deg', 'weather_id', 'weather_main']]
rawdata
```

```python
#konversi object tanggal menjadi date time
rawdata['dt_iso'] = pd.to_datetime(rawdata['dt_iso'])
```

```python
rawdata.info()
```

```python
#memfilter data tahun yang lebih besar sama dengan 1 januari 2020
mask = ((rawdata['dt_iso'] >= '2019-01-01') & (rawdata['dt_iso'] <= '2019-01-31'))
```

```python
data19 = rawdata.loc[mask]
```

```python
data19
```

```python
import matplotlib.pyplot as plt 

x_axis = data19['dt_iso']
y_axis = data19['temp']

# plotting a line plot after changing it's width and height 
f = plt.figure() 
f.set_figwidth(35) 
f.set_figheight(15) 
  
plt.plot(x_axis, y_axis)
plt.title('Sebaran Temperatur dari 1 Januari 2019 s.d 31 Januari 2019')
plt.xlabel('Waktu')
plt.ylabel('Kelembapan')
plt.show()
```
